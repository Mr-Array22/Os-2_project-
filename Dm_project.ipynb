```python
#visualization
fig=plt.hist(dataset.price)
plt.xlabel('price')
```

```python
fig=plt.figure(figsize=(12,6))
android_version=fig.add_subplot(121)
price=fig.add_subplot(122)

price.hist(dataset.price,bins=10)
android_version.hist(dataset.android_version,bins=10)

price.set_title('Histogram of Price')
android_version.set_title('Histogram of android_version')

price.set_xlabel(' Price')
android_version.set_xlabel(' android_version')



plt.show()

```
